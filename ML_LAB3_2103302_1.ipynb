```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Setting size(n) = 100 and probability of success(p) = 0.4
n = 100
p1 = 0.4
```

```python
# Generating 100 points by Bernoulli distribution which are i.i.d. from p = 0.4
points = np.random.choice([0, 1], size=n, p=[1-p1, p1])
print(points)
```

```python
# Counting number of 0s and 1s
n0 = n - np.sum(points)
n1 = np.sum(points)
# Sum of 1s = Number of 1s
print(f'Number of 0s = {n0}')
print(f'Number of 1s = {n1}')
```

```python
# Calculating likelihood as function of p
def likelihood(p, n0, n1):
  return ((1 - p)**n0) * (p**n1)
```

```python
# Calculating log of likelihood
def log_likelihood(p, n0, n1):
  return n0*np.log(1 - p) + n1*np.log(p)
```

```python
# Calculating estimator (mean of Bernoulli distribution)
def estimator(points, n):
  sum = 0;
  for i in range(n):
    sum += (points[i] / n)
  return sum
```

```python
# Printing everything
print("Likelihood, L(p) = {}".format(likelihood(p1, n0, n1)))
print("Log of Likelihood, l(p) = {}".format(log_likelihood(p1, n0, n1)))
print("Estimator, p^ = {}".format(estimator(points, n)))
```

```python
# Plotting Likelihood Graph
L_X = np.arange(0.000, 1.001, 0.001)
L_Y = np.zeros(len(L_X), float)
for i in range(len(L_X)):
  L_Y[i] = likelihood(L_X[i], n0, n1)
plt.plot(L_X, L_Y)
plt.title("Likelihood Plot")
plt.xlabel("p")
plt.ylabel("Likelihood")
plt.show()
```

```python
# Plotting log_Likelihood Graph
l_X = np.arange(0.000, 1.001, 0.001)
l_Y = np.zeros(len(l_X), float)
for i in range(len(l_X)):
  l_Y[i] = log_likelihood(l_X[i], n0, n1)
plt.plot(l_X, l_Y)
plt.title("log of Likelihood Plot")
plt.xlabel("p")
plt.ylabel("ln(Likelihood)")
plt.show()
```
