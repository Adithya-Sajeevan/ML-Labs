```python
import numpy as np
import pandas as pd
from math import sqrt
from matplotlib import pyplot as plt
from sys import maxsize
```

```python
file = "oldFaithfulGeyserDataset.csv"
df = pd.read_csv(file)
Data = np.array(df)
Data = Data.T[1:3].T
# Old Faithful Geyser Dataset
```

```python
k, n = 2, len(Data)
Dict = {}
for j in range(k):
  Dict[j] = [];
for i in range(n):
  Dict[i % k].append(Data[i])
```

```python
move = True
itr = 0
Color = np.random.rand(k, 3)
print(f'itr:{itr}->')
for j in range(k):
  A = np.array(Dict[j])
  X, Y = A[:,0], A[:,1]
  RGB = (Color[j][0], Color[j][1], Color[j][2])
  plt.scatter(X, Y, c=RGB)
plt.show()
while move == True:
  itr += 1
  print(f'itr:{itr}->')
  C = np.empty((k, 2))
  for j in range(k):
    A = np.array(Dict[j])
    C[j][0] = np.sum(A.T[0]) / len(A)
    C[j][1] = np.sum(A.T[1]) / len(A)
  print(f'Centroid = {C}')
  _Dict = {}
  for j in range(k):
    _Dict[j] = []
  for i in range(n):
    data = Data[i]
    min_dist = maxsize
    index = 0
    for j in range(k):
      dist = sqrt((data[0] - C[j][0])**2 + (data[1] - C[j][1])**2)
      if dist < min_dist:
        index = j
        min_dist = dist
    _Dict[index].append(data)
  print("New Classification:")
  print(_Dict)
  for j in range(k):
    if np.array_equal(_Dict[j], Dict[j]) == True:
      move = False
  for j in range(k):
    A = np.array(Dict[j])
    X, Y = A.T[0], A.T[1]
    plt.scatter(X, Y, c=(Color[j][0], Color[j][1], Color[j][2]))
  X, Y = C[:,0], C[:,1]
  plt.scatter(X, Y, s=100, c="red", marker='x')
  plt.show()
  Dict = _Dict
  print()
```
